# What is a Financial Budget?


A financial budget is a comprehensive plan that outlines an individual's, company's, or organization's expected income and expenses over a specific period, typically a fiscal year. It serves as a roadmap for managing financial resources, setting financial goals, and ensuring that resources are allocated efficiently to achieve those objectives.

```python
import pandas as pd

# Assuming 'df' is your existing DataFrame
new_row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}

# Use concat to add the new row
data = pd.concat([data, pd.DataFrame([new_row])], ignore_index=True)

print(data)
```

```python
# Top 10 ministries/departments by budget allocation
top_10 = data.nlargest(10, 'Fund allotted(in ₹crores)')
print(top_10)
```

```python
# Visualize top 10 ministries/departments
top_10.plot(kind='bar', x='Department /Ministry', y='Fund allotted(in ₹crores)', figsize=(12, 6), rot=45)
plt.title('Top 10 Ministries/Departments by Budget Allocation')
plt.xlabel('Ministry/Department')
plt.ylabel('Allocation (₹ crores)')
plt.tight_layout()
plt.show()
```

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)')
```

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```
